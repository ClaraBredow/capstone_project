```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import missingno as msno
```

```python
df = pd.read_csv('data_strube/openweather_droppedcolumns_2021.csv')
```

```python
# simplify weatherdata for now
df.drop(['dt', 'dt_iso', 'timezone','feels_like', 'clouds_all', 'weather_id',
       'weather_main', 'weather_description', 'weather_icon'], 
                         axis=1, 
                         inplace=True
                         )
```

```python
df.columns
```

```python
df.rename(columns={'city_name': 'station_location'}, inplace=True)
#remove the numbers from the fieldid
```

```python
df['station_location'] = df['station_location'].replace(r'VierhÃ¶fen', r'Vierhoefen', regex=True)
```

```python
df_grouped = df.groupby(['station_location', 'month'], as_index=False).mean()
df_grouped
```

```python
# drop day columns and remove duplicate rows
df_grouped.drop_duplicates(keep='first', inplace=True)
df_grouped.shape
```
